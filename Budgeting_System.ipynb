```python
import numpy as np
import matplotlib.pyplot as plt

name = input("What is your name? ")
age = int(input("What is your age? "))
salary = int(input("Enter your salary: "))

# Define fixed percentages
rent_pct = 0.30
food_pct = 0.15
electricity_pct = 0.05
water_pct = 0.03
transport_pct = 0.08
necessities_pct = 0.10

# Healthcare percentage function
def healthcare_pct(age):
    if age >= 70:
        return 0.20
    elif age >= 60:
        return 0.15
    elif age >= 40:
        return 0.10
    else:
        return 0.05

# Calculate healthcare percentage based on age
healthcare_percentage = healthcare_pct(age)

# Create percentages array with fixed + healthcare
percentages = np.array([
    rent_pct, food_pct, electricity_pct, water_pct,
    healthcare_percentage, transport_pct, necessities_pct
])

# Calculate "Other" = remainder
other_pct = 1 - np.sum(percentages)

# Add "Other"
percentages = np.append(percentages, other_pct)

# Expense names
expenses = [
    'Rent', 'Food', 'Electricity', 'Water',
    'Healthcare', 'Transport', 'Necessities', 'Other'
]

# Calculate actual amounts
amounts = percentages * salary


# PRINT RESULTS TO USER
print(f"\nBudget Breakdown for {name} (Age: {age}):")
print("=" * 40)
for i in range(len(expenses)):
    print(f"{expenses[i]}: sh{amounts[i]:,.2f} ({percentages[i]*100:.1f}%)")

print("=" * 40)
print(f"Total income: sh{salary:,.2f}")
print(f"Total allocated: sh{np.sum(amounts):,.2f}")


# PIE CHART
plt.figure(figsize=(8, 8))
plt.pie(amounts, labels=expenses, autopct="%1.1f%%", startangle=140)
plt.title(f"Expense Distribution Pie Chart for {name}")
plt.show()

# BAR CHART
plt.figure(figsize=(10, 6))
plt.bar(expenses, amounts)
plt.xlabel("Expense Category")
plt.ylabel("Amount (sh)")
plt.title(f"Monthly Budget Bar Chart for {name}")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
